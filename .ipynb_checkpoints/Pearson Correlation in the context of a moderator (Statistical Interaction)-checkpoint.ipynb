#### import necessary libraries

```python
import pandas as pd
import numpy as np
import scipy.stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### read in the data

```python
data1 = pd.read_csv('gapminder2.csv', low_memory=False)
```

#### remove unnecessary columns and make a copy of the subdata

```python
data2 = data1[["continent", "country", "breastcancerper100th", "urbanrate", "incomeperperson", "breastcancernbdeaths"]]
data = data2.copy()
```

#### remove missing values

```python
data= data.replace(0, np.NaN)
data = data.dropna()
```

#### Change the data type for chosen variables

```python
data['breastcancerper100th'] = pd.to_numeric(data['breastcancerper100th'])
data['incomeperperson'] = pd.to_numeric(data['incomeperperson'])
data['urbanrate'] = pd.to_numeric(data['urbanrate'])
```

#### Request Pearson Correlation

```python
print ("association between urbanrate and breast cancer cases")
print (scipy.stats.pearsonr(data['urbanrate'], data['breastcancerper100th']))
```

#### Add a moderator variable to the data

```python
def incomegrp (row):
   if row['incomeperperson'] <= 744.239:
      return 1
   elif row['incomeperperson'] <= 9425.326 :
      return 2
   elif row['incomeperperson'] > 9425.326:
      return 3
```

```python
data['incomegrp'] = data.apply (lambda row: incomegrp (row),axis=1)
```

```python
chk1 = data['incomegrp'].value_counts(sort=False, dropna=False)
print(chk1)
```

#### create sub-datasets for each value of incomegrp

```python
sub1=data[(data['incomegrp']== 1)]
sub2=data[(data['incomegrp']== 2)]
sub3=data[(data['incomegrp']== 3)]
```

#### Run Pearson Correlation for every subset

```python
print ('association between urbanrate and breast cancer cases for LOW income countries')
print (scipy.stats.pearsonr(sub1['urbanrate'], sub1['breastcancerper100th']))
```

```python
print ('association between urbanrate and internetuserate for MIDDLE income countries')
print (scipy.stats.pearsonr(sub2['urbanrate'], sub2['breastcancerper100th']))
```

```python
print ('association between urbanrate and internetuserate for HIGH income countries')
print (scipy.stats.pearsonr(sub3['urbanrate'], sub3['breastcancerper100th']))
```

#### Create plots to vizualize the associations

```python
scat1 = sns.regplot(x="urbanrate", y="breastcancerper100th", data=sub1)
plt.xlabel('Urban Rate')
plt.ylabel('Breast cancer cases')
plt.title('Scatterplot for the association between urban rate and breast cancer cases for LOW income countries')
print (scat1)
```

```python
scat2 = sns.regplot(x="urbanrate", y="breastcancerper100th", data=sub2)
plt.xlabel('Urban Rate')
plt.ylabel('Breast Cancer Cases')
plt.title('Scatterplot for the association between urban rate and breast cancer cases for MIDDLE income countries')
print (scat2)
```

```python
scat3 = sns.regplot(x="urbanrate", y="breastcancerper100th", data=sub3)
plt.xlabel('Urban Rate')
plt.ylabel('Breast Cancer Cases')
plt.title('Scatterplot for the association between urban rate and breast cancer cases for HIGH income countries')
print (scat3)
```
